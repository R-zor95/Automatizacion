## Creando un camino a partir de un genoma

Para cada casilla, el genoma nos designa una dirección, en la que un robot intentaría moverse una vez que ha llegado a esa casilla.

```python
mapa1.create_camino()
ejemplo = mapa1.list_caminos[0]
ag.calculate_traits(ejemplo, ejemplo.dict_genes)
lab.calculate_performances(ejemplo)
ejemplo.draw_directions()
```

Si el robot sigue estas direcciones no encontraria un camino como tal , aqui las restricciones:


- Si la dirección de la casilla le dice que intente atravesar una pared, el robot intentará otra dirección aleatoria.

- El robot tendrá una cantidad limitada de pasos.

```python
ejemplo.draw_directions()
ejemplo.draw_path()
```

####   Implementando el algoritmo genético

- Cuando el robot choque con una pared, el camino recibirá una pequeña penalización.

- Si hay dos casillas que se apuntan mutuamente, el robot quedará atrapado, por lo que también penalizamos volver a la casilla de la que se partió en el turno anterior.

- Queremos llegar a la salida de la derecha, por lo que beneficiaremos a los caminos que dejen al robot más a la derecha.

- Si un camino conduce a la salida, le daremos una gran bonificación para no perderlo

```python
lab.calculate_fitness(ejemplo)
ejemplo.fitness
```
